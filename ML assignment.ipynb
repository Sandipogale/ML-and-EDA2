Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent them?

In Machine Learning, building an accurate predictive model depends on how well the model learns patterns from the training data and how well it generalizes to unseen data. Two common problems that occur during model training are overfitting and underfitting. Both issues affect the performance of the model, but in opposite ways. Understanding them requires knowledge of the bias–variance tradeoff, along with techniques such as cross-validation and regularization.
Overfitting
Overfitting happens when a model learns the training data too well, including noise, random fluctuations, and unnecessary patterns. As a result, the model performs extremely well on training data but fails to generalize to new, unseen data.

Characteristics of Overfitting

Very high training accuracy but low test/validation accuracy

The model becomes overly complex

High variance (model predictions change a lot with small changes in input)

Example

If you draw a very complicated curve through every point on a scatterplot to achieve zero error, the model will memorize the data instead of learning the underlying pattern.

Underfitting
Underfitting occurs when the model is too simple to learn the underlying patterns in the data. It does not perform well on training data or test data.

Characteristics of Underfitting

Low training accuracy and low test accuracy

Model is too simple to capture the relationships in data

High bias (model makes strong assumptions and oversimplifies the data)

Example

Using a straight-line model (linear regression) to fit a dataset that clearly has a curved pattern.

Bias–Variance Tradeoff Bias refers to errors due to overly simple assumptions.
Variance refers to sensitivity to small changes in the training data.

A good ML model maintains a balance between bias and variance so that it neither memorizes the training data nor oversimplifies the problem.

How to Detect Overfitting and Underfitting Detecting Overfitting
Training accuracy is very high, validation/test accuracy is significantly lower.

Model complexity increases without improvement in test performance.

Learning curve shows large gap between training and validation loss.

Detecting Underfitting

Both training and test accuracy are low.

Increasing model complexity improves performance.

Learning curve shows that training and validation errors remain high.

How to Prevent Overfitting a) Cross-Validation (especially k-Fold Cross Validation)
Instead of using a single train-test split, k-fold CV divides data into k parts and trains the model repeatedly. This ensures the model is tested on different subsets, helping detect and reduce overfitting.

b) Regularization Techniques

Regularization adds penalties to the model to discourage overly complex patterns.

Common regularization methods:

L1 Regularization (Lasso) – shrinks less important feature weights to zero

L2 Regularization (Ridge) – reduces the magnitude of weights

Elastic Net – combination of L1 and L2

These methods prevent models from becoming overly complex.

c) Early Stopping

When training neural networks or gradient boosting models, stop training once validation loss starts increasing even if training loss is decreasing.

d) Pruning (for Decision Trees)

Limiting tree depth, removing branches that don’t contribute significantly, and controlling minimum samples per split.

Missing values are a common problem in real-world datasets. They occur due to various reasons such as human errors during data entry, sensor failures, skipped survey questions, or incomplete records. Handling missing values properly is crucial because they can affect the accuracy of the model, distort statistical results, and weaken the overall conclusions. Data preprocessing therefore includes identifying, understanding, and treating missing values appropriately. There are multiple strategies to handle missing data, and the choice depends on the type of data, the amount of missingness, and the importance of the missing variable.
Below are three widely used methods to handle missing values, along with explanations and examples.

Deletion Methods (Removing Missing Data)
Deletion means removing rows or columns that contain missing values. This method is simple but should be used carefully because it can lead to loss of useful data.

a) Listwise Deletion (Removing Entire Rows)

If a row contains even one missing value, the entire row is removed.

When to use

When the dataset is large

Missing values are very few

Missingness is random (no pattern)

When the column is not important for analysis

a) Mean Imputation (for numerical data)

Question 4:What is an imbalanced dataset? Describe two techniques to handle it?

In many real-world machine learning problems, the distribution of classes is not equal. Some classes may have a large number of instances while others have very few. Such uneven class distribution creates what is known as an imbalanced dataset. Handling imbalance correctly is very important because most ML algorithms assume that classes are roughly balanced. When this assumption is violated, the model becomes biased toward the majority class and performs poorly on the minority class.
What is an Imbalanced Dataset?
A dataset is said to be imbalanced when the number of observations in each class is not proportional. For example, in fraud detection, only 1 out of 1000 transactions may be fraudulent. So the class distribution might look like:

Class 0 (Legitimate): 99%

Class 1 (Fraud): 1%

Even a useless model that predicts “Not Fraud” for every record will give 99% accuracy, but it fails to detect fraud. Thus, accuracy becomes misleading, and specialized techniques are required.

Why is Imbalance a Problem?
Model becomes biased toward majority class

Minority class (often the more important class) gets ignored

Metrics like accuracy become meaningless

Poor generalization on real-world data

Harder for the model to learn patterns of minority class

Thus, handling imbalance is essential for reliable predictions.

Techniques to Handle Imbalanced Datasets
Below are two widely used and effective methods — one data-level and one algorithm-level — with both theoretical and practical explanations.

Technique 1: Random Over/Undersampling (Data-Level Method) A. Oversampling (Increasing Minority Class Samples) ✓ Theory

Oversampling involves duplicating or synthetically generating more samples of the minority class so that it becomes balanced with the majority class.

✓ Common Approaches

Random Oversampling Simply duplicates existing samples of the minority class.

✓ Why it Works

Balances class distribution

Prevents the model from being biased

Simple and fast

✓ Limitation

Can cause overfitting because it repeats the same rows.

B. Undersampling (Reducing Majority Class Samples) ✓ Theory

Undersampling removes some portion of the majority class so that both classes have similar counts.

✓ Why it Works

Fast and simple

Useful when majority class has too many samples

Reduces training time

✓ Limitation

Loss of useful information

May remove important patterns

```python
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter
import pandas as pd
import numpy as np

# Create a sample imbalanced DataFrame for demonstration
np.random.seed(42)
n_majority = 1000
n_minority = 50

# Majority class
X_majority = np.random.rand(n_majority, 2)
y_majority = np.zeros(n_majority)

# Minority class
X_minority = np.random.rand(n_minority, 2) + 0.5 # Shift for distinction
y_minority = np.ones(n_minority)

X = np.vstack((X_majority, X_minority))
y = np.hstack((y_majority, y_minority))

df = pd.DataFrame(X, columns=['feature1', 'feature2'])
df['target'] = y

print("Before:", Counter(df['target']))

X = df.drop('target', axis=1)
y = df['target']


# Oversampling
ros = RandomOverSampler(random_state=42)
X_over, y_over = ros.fit_resample(X, y)
print("After Oversampling:", Counter(y_over))

# Undersampling
rus = RandomUnderSampler(random_state=42)
X_under, y_under = rus.fit_resample(X, y)
print("After Undersampling:", Counter(y_under))
```

Technique 2: SMOTE (Synthetic Minority Oversampling Technique) A. Theory

SMOTE is an advanced oversampling technique that creates synthetic (not duplicate) samples of the minority class.

✓ How SMOTE Works

Picks a sample from the minority class

Finds its k-nearest neighbors

Creates new synthetic points along the line between the sample and its neighbors

This increases diversity and reduces overfitting

✓ Advantages of SMOTE

Generates new, realistic samples

Reduces model bias

Works well with many ML algorithms

✓ Limitations

Can create noisy samples

Not ideal for high-dimensional data without cleaning

Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.

Why Feature Scaling Is Important
Feature scaling ensures that all numerical features are on a similar scale. It is important because:

Prevents Features With Large Values From Dominating
Algorithms like KNN, K-Means, SVM use distance-based calculations. If one feature ranges from 1–1000 and another from 1–10, the large-range feature will dominate the distance → incorrect model behavior.

Speeds Up Gradient Descent Convergence
Algorithms like:

Linear Regression

Logistic Regression

Neural Networks

use gradient descent. If features are not scaled, the cost function becomes stretched → the algorithm takes longer to converge.

Improves Model Accuracy & Stability
Scaled features often result in:

better optimization

more stable training

reduced chances of exploding/vanishing gradients

Aspect	Min-Max Scaling	Standardization (Z-score)
Formula	(x – min) / (max – min)	(x – mean) / standard deviation
Range	0 to 1 (or custom range)	No fixed range, mean = 0, sd = 1
Sensitive to outliers	YES — highly sensitive	Less sensitive
Good for	Neural networks, KNN, K-Means, distance-based ML	Linear models, Logistic regression, SVM, PCA
When to use	When features have similar, bounded ranges	When data follows a normal distribution

Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?

In machine learning, many datasets contain categorical features such as gender, color, location, product type, education level, etc. Since most ML algorithms cannot directly understand text labels, these categorical variables must be converted into numerical format. Two of the most widely used techniques for this purpose are Label Encoding and One-Hot Encoding. Although both methods convert categories into numeric representations, they differ significantly in terms of approach, interpretation, use cases, and the impact they create on the learning algorithms. Below is a detailed comparison of both techniques and the scenarios where each one is preferred.
Label Encoding (Theory)
Label Encoding is a method in which each category in a categorical column is assigned a unique integer value. The technique replaces text-based categories with numbers such as 0, 1, 2, 3, and so on. For example, a column containing size values:

Small

Medium

Large

may be encoded as: Small → 0 Medium → 1 Large → 2

1.1. How Label Encoding Works

The encoder scans all unique categories in a feature and assigns each one an integer based on alphabetical order or its internal mapping logic. This transformation is simple, fast, and does not increase the number of columns.

1.2. Advantages of Label Encoding

Very simple and memory-efficient, since it converts categories into a single numerical column.

Works extremely well for ordinal data, where categories naturally follow a meaningful order.

Ideal for algorithms such as Decision Trees, Random Forests, and Gradient Boosting, because these models split nodes based on conditions and do not misinterpret the numbering as mathematical relationships.

1.3. Limitations of Label Encoding

One major issue with label encoding is that it introduces an artificial order among the categories. For example, if we encode colors as: Red → 0 Blue → 1 Green → 2

The model may wrongly assume that:

Green (2) is “greater than” Blue (1)

Blue (1) is “greater than” Red (0)

This is not meaningful because colors do not have any natural ranking. Thus, label encoding can mislead algorithms like KNN, SVM, Logistic Regression, and Linear Regression, which treat numeric differences as actual distances or magnitudes.

One-Hot Encoding (Theory)
One-Hot Encoding is a method that converts each category into a separate binary (0/1) column. For example, the “Color” feature with three categories—Red, Blue, Green—becomes:

Red Blue Green 1 0 0 0 1 0 0 0 1

Each row gets a “1” under the category it belongs to and “0” under the others.

2.1. How One-Hot Encoding Works

The encoder examines a categorical column and creates a new binary feature for each unique category. Thus, if a column has N unique categories, the method creates N new columns. This encoding removes any possibility of the model assuming an incorrect mathematical order.

2.2. Advantages of One-Hot Encoding

Prevents models from assuming a ranking among categories.

Works extremely well for nominal data, where no natural order exists.

Suitable for distance-based algorithms such as KNN, as well as linear models like SVM, Logistic Regression, Linear Regression, because it keeps relationships neutral.

Produces more accurate and unbiased results when used correctly.

2.3. Limitations of One-Hot Encoding

Increases dimensionality significantly, especially for features with many unique values (e.g., thousands of cities or product categories).

High dimensionality leads to more memory usage, slower training, and risk of the curse of dimensionality.

Not ideal for very large datasets containing high-cardinality features.

When to Prefer Label Encoding
You should use label encoding when the categorical data is ordinal, meaning the categories have an inherent order or ranking. Examples include:

Education Level: High School < Bachelor < Master < PhD

Size: Small < Medium < Large

Customer Satisfaction: Poor < Average < Good < Excellent

In these cases, the numeric representation helps the model understand the progression or hierarchy among categories. Additionally, label encoding works well for tree-based models because these models split data boundaries without relying on numeric magnitude.

When to Prefer One-Hot Encoding
One-Hot Encoding should be used when the categorical data is nominal, meaning the categories do not have an inherent order. Examples include:

Gender: Male, Female

Country: India, USA, Japan

Color: Red, Blue, Green

App Category: Games, Business, Education

In such cases, applying label encoding could mislead the model into interpreting relationships that do not exist. One-hot encoding is especially beneficial when using algorithms that depend heavily on numerical scale, such as:

KNN (distance-based)

SVM

Logistic Regression

Linear Regression

Neural Networks

These algorithms perform better when categorical data is represented using unbiased binary variables.

Question 7: Google Play Store Dataset

highest/lowest average ratings, and what could be the possible reasons?

```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the Google Play Store dataset
# Replace with your actual CSV path
file_path = r"C:\Users\Sandip\Downloads\googleplaystore.csv"
df = pd.read_csv(file_path)

# Display first few rows
print("First 5 rows of dataset:")
print(df.head())

# 2. Data cleaning
# Remove invalid ratings if any (non-numeric values or NaN)
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df = df.dropna(subset=['Rating'])

# 3. Analyze average rating by Category
category_rating = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)
print("\nAverage Rating by Category:")
print(category_rating)

highest_category = category_rating.idxmax()
lowest_category = category_rating.idxmin()
print(f"\nCategory with highest average rating: {highest_category}")
print(f"Category with lowest average rating: {lowest_category}")

# 4. Visualization
plt.figure(figsize=(12,6))
sns.barplot(x=category_rating.index, y=category_rating.values)
plt.xticks(rotation=90)
plt.title("Average Rating by App Category")
plt.ylabel("Average Rating")
plt.xlabel("Category")
plt.show()
```

consistently cheaper/premium, and why?

```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the Flight Price dataset
file_path = r"C:\Users\Sandip\Downloads\flight_price.csv"  # Replace with your CSV path
df = pd.read_csv(file_path)

# Display first few rows and columns to understand structure
print("Columns in dataset:", df.columns)
print(df.head())

# 2. Detect columns for 'Days Left' and 'Price'
# We'll try common variants
possible_days_columns = ['days_left', 'DaysLeft', 'Days To Departure', 'Days_To_Departure']
possible_price_columns = ['price', 'FlightPrice', 'Cost']

days_col = next((col for col in possible_days_columns if col in df.columns), None)
price_col = next((col for col in possible_price_columns if col in df.columns), None)

if not days_col or not price_col:
    raise ValueError("Could not find suitable columns for Days Left or Price. Check your CSV column names.")

# 3. Scatter plot: Price vs Days Left
plt.figure(figsize=(10,6))
sns.scatterplot(x=days_col, y=price_col, data=df)
plt.title("Flight Price vs Days Left Until Departure")
plt.xlabel("Days Left")
plt.ylabel("Price")
plt.show()

# Optional: Line plot to check trends
df_sorted = df.sort_values(by=days_col)
plt.figure(figsize=(10,6))
sns.lineplot(x=days_col, y=price_col, data=df_sorted)
plt.title("Flight Price Trend vs Days Left")
plt.xlabel("Days Left")
plt.ylabel("Price")
plt.show()

source_col = next((col for col in ['Source', 'From', 'DepartureCity'] if col in df.columns), None)
dest_col = next((col for col in ['Destination', 'To', 'ArrivalCity'] if col in df.columns), None)
airline_col = next((col for col in ['Airline', 'Carrier'] if col in df.columns), None)

if not source_col or not dest_col or not airline_col:
    raise ValueError("Could not find suitable columns for Source, Destination, or Airline.")

route_df = df[(df[source_col] == 'Delhi') & (df[dest_col] == 'Mumbai')]

# Average price per airline
airline_price = route_df.groupby(airline_col)[price_col].mean().sort_values()
print("\nAverage prices per airline (Delhi-Mumbai):")
print(airline_price)

# Bar plot
plt.figure(figsize=(12,6))
sns.barplot(x=airline_price.index, y=airline_price.values)
plt.xticks(rotation=45)
plt.title("Average Flight Prices by Airline (Delhi-Mumbai)")
plt.ylabel("Average Price")
plt.xlabel("Airline")
plt.show()

best_booking_window = df_sorted[price_col].idxmin()
print(f"\nRecommended best booking window: Around {df_sorted.iloc[best_booking_window][days_col]} days before departure (lowest price observed).")
```
